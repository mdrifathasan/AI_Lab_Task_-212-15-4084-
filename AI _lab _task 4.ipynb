```python
import pandas as pd

df = pd.read_csv('/content/PriceDataSet.csv')

df['price'] = pd.to_datetime(df['price'])

print(df.to_string())
```

```python
df.dropna(subset=['price'], inplace = True)
```

```python
df.loc[7, 'Duration'] = 10
```

```python
for x in df.index:
  if df.loc[x, "Duration"] > 10:
    df.loc[x, "Duration"] = 12
```
