```python
df = pd.read_csv('/content/PriceDataSet (2).csv')
```

```python
new_df = df.dropna()
print(df)
print(new_df)
```

```python
new_df = df.fillna(50)
print(new_df)
```

```python
x = df["weight"].mean()
print('Mean is ', x)
new_df = df.fillna(x)
print(new_df)
```

```python
x = df["weight"].median()
print('Median is ', x)
new_df = df.fillna(x)
print(new_df)
```

```python
x = df["weight"].mode()[0]
print('Mode is ', x)
new_df = df.fillna(x)
print(new_df)
```
